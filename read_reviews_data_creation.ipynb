```python
# Function to read and create master dataframe: 
def read_tagged_files(path): 
    tagged_files = os.listdir(path)
    tagged_files.remove('amazon_reviews_office_0_all.xlsx')
    print(tagged_files)
    master_df= pd.DataFrame()
    for file in tagged_files: 
        print(file)
        if file != '.ipynb_checkpoints':
            df = pd.read_excel(os.path.join(path, file))
            df = df[['review_id','full_review',
           'sentence','rating_managment_explicit',
           'disagreement_with_ratings', 'wrong_buying', 'zero_star',
           'wish_I_read_reviews ' ]]
            master_df = pd.concat([master_df,df], ignore_index=True)
    return master_df 

def one_hot_encode(cnt): 
    if cnt>=1: 
        return 1
    else: 
        return 0 
# for each label create dataset
def create_label_dataset(df, label): 
    # seperate review_id, full_review: 
    # take only review id and full review
    df_1 = df[['review_id', 'full_review']] 
    df_1 = df_1.drop_duplicates()
    df_2 = df.groupby('review_id')[label].sum().reset_index()
    label_df = df_1.merge(df_2, on = 'review_id', how = 'inner')
    label_df[f'{label}_ohe']= label_df[label].apply(lambda x:one_hot_encode(x))
    return label_df
```

```python
# read_reviews
master_df = read_tagged_files(os.path.join(os.getcwd(), 'tagged_data'))
rating_managment_explicit= create_label_dataset(master_df, 'wish I read reviews ')
pd.set_option('display.max_rows', None)
rating_managment_explicit.to_csv('/home/ec2-user/SageMaker/read_reviews/read_reviews_v1.csv', index = None)
```
